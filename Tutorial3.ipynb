```python
from firebase import firebase
FBconn = firebase.FirebaseApplication('https://tutorial3-99b15-default-rtdb.firebaseio.com/', None)

def get_all_words():
    """Retrieving all words from the database"""
    return FBconn.get('/Words/', None)

def update_word_count(word, new_count):
    """Update a counter to a specific word - use put to make the word the key"""
    return FBconn.put('/Words/', word, new_count)

def delete_word(word):
    """Word deletion"""
    return FBconn.delete('/Words/', word)

def display_all_words():
    """Showing all words and their masses"""
    words_data = get_all_words()
    if words_data:
        print("\n--- Current Word Counts ---")
        for word, count in words_data.items():
            print(f"{word}: {count}")
    else:
        print("\nNo words found in database.")

while True:
    print("\nCommon Words Counter Menu:")
    print("1. Add single word")
    print("2. Add text for analysis")
    print("3. Update word count (Manually)")
    print("4. Delete word")
    print("5. View all Words")
    print("6. Exit")

    choice = input("\nEnter your choice (1-6): ")

    if choice == '1':
        word = input("Enter a word: ").lower().strip()
        current_val = FBconn.get('/Words/', word)
        if current_val is None:
            update_word_count(word, 1)
        else:
            update_word_count(word, current_val + 1)
        print(f"Word '{word}' updated.")

    elif choice == '2':
        text = input("Enter text for analysis: ").lower()
        words_list = text.split()
        for w in words_list:
            w = w.strip('.,!?:;')
            current_val = FBconn.get('/Words/', w)
            if current_val is None:
                update_word_count(w, 1)
            else:
                update_word_count(w, current_val + 1)
        print("Text analyzed and uploaded!")

    elif choice == '3':
        display_all_words()
        word = input("Which word to update manually? ").lower()
        new_count = int(input("Enter new count: "))
        update_word_count(word, new_count)

    elif choice == '4':
        display_all_words()
        word_to_del = input("Enter the word to delete: ").lower()
        delete_word(word_to_del)
        print(f"Word '{word_to_del}' deleted.")

    elif choice == '5':
        display_all_words()

    elif choice == '6':
        print("Exiting program...")
        break
```
