Gemini: Creating a prompt
This rich example shows how you can create and configure complex prompts for Gemini. It assumes that you've already created an API key at https://aistudio.google.com/app/apikey and added it to your Colab secrets as GOOGLE_API_KEY (see the "Connecting to Gemini" snippet).

```python
import google.generativeai as genai

genai.configure(api_key="YOUR API KEY")

model = genai.GenerativeModel("gemini-2.5-flash")

system_instruction = """
You are the AI Botanist Assistant for 'OnlyPlants' - a smart IoT plant care dashboard specifically designed for Orchid (Phalaenopsis) growers.
Your job is to help users analyze plant health, suggest care routines based on real-time environmental data (humidity, temperature, light, soil moisture),
and explain app features like Gamification (earning points for proper care) and AI Image Analysis for disease detection (like Botrytis or root rot).
Keep your answers brief, informative, encouraging, and tailored to orchid precision agriculture.
"""

print("Welcome to OnlyPlants AI Assistant! 🌿 (Type 'quit' or 'exit' to stop)")

while True:
    user_question = input("\nAsk the AI Botanist: ")

    if user_question.lower() in ['quit', 'exit', 'יציאה', 'צא']:
        print("Goodbye! Keep your orchids blooming!")
        break

    full_prompt = f"{system_instruction}\n\nUser Question: {user_question}"

    try:
        response = model.generate_content(full_prompt)
        print("\n🤖 OnlyPlants AI:", response.text)
    except Exception as e:
         print("\n❌ Error communicating with AI:", e)
```

```python
import nltk
from nltk.chat.util import Chat, reflections

# Download necessary NLTK data (run once)
nltk.download('punkt')
nltk.download('wordnet')

# Define patterns and responses for the OnlyPlants chatbot based on the project index
patterns = [
    # 1. Greetings
    (r'hi|hello|hey',
     ['Hello! Welcome to the OnlyPlants care assistant.',
      'Hi there! How are your orchids blooming today?',
      'Hey! Ready to check on your smart garden?']),

    (r'how are you\??',
     ['I am feeling green and fully charged! How can I help your orchids?',
      'Doing great, just analyzing some soil telemetry!']),

    # 2. Watering, Moisture, and Irrigation
    (r'(.*)(water|watering|moisture|irrigation|dry|wet)(.*)',
     ['OnlyPlants monitors your soil moisture levels in real-time. If it drops below optimal levels, you will get a notification!',
      'You can check the active moisture status of your orchid directly on the main dashboard screen.',
      'Proper irrigation is critical to prevent root rot. Ensure the moisture sensor is properly inserted.']),

    # 3. Light, Sunlight, and Photosynthesis
    (r'(.*)(light|sun|sunlight|shade|dark|photosynthesis)(.*)',
     ['Our IoT light sensors track daily light exposure to ensure optimal photosynthesis.',
      'If the app shows "Low Light" warnings, consider moving your OnlyPlants device closer to a window, but avoid direct sunlight which can burn the leaves.']),

    # 4. Temperature and Humidity
    (r'(.*)(temperature|heat|cold|humidity|greenhouse)(.*)',
     ['Phalaenopsis orchids thrive in specific temperature and humidity ranges. Check the dashboard for real-time readings from the greenhouse.',
      'High humidity without proper ventilation can lead to fungal diseases like Botrytis.']),

    # 5. Disease, Rot, and Pathogens
    (r'(.*)(disease|rot|pathogen|botrytis|viral)(.*)',
     ['Our AI image analysis can detect early signs of diseases like root rot or Botrytis. Please upload a clear photo of the affected leaf.',
      'Viral diseases often present as unique patterns on the leaves. Use the AI diagnosis tool for a preliminary check.']),

     # 6. Leaf, Morphology, and Physiology
    (r'(.*)(leaf|leaves|morphology|physiology)(.*)',
     ['The leaf is a critical indicator of plant health. Any changes in morphology, like drooping or spots, should be monitored.',
      'Our AI models analyze leaf images to assess the physiological state of the orchid.']),

    # 7. Gamification and Points
    (r'(.*)(points|score|game|badge|level|leaderboard)(.*)',
     ['Every time you complete a care task on time (like watering or adjusting light), you earn OnlyPlants points!',
      'Keep your plant health score high to level up, unlock virtual badges, and top the community leaderboard!']),

    # 8. Hardware, Sensors, and IoT
    (r'(.*)(sensor|iot|connect|wifi|offline|real-time)(.*)',
     ['The system provides real-time data from the sensors to the cloud. If a sensor is offline, please check your Wi-Fi connection.',
      'Ensure the soil probe is deeply inserted into the dirt for accurate moisture readings.']),

     # 9. AI, CNN, and Image Detection
    (r'(.*)(ai|image|detection|cnn|dataset|smart|precision)(.*)',
     ['We use advanced Computer Vision Pipelines (CNNs) trained on annotated imagery to detect orchid diseases with high precision.',
      'The smart detection system helps in early identification of pathogens, saving your orchids before the disease spreads.']),

     # Fallback
     (r'(.*)',
     ['I am not sure I understand. You can ask me about watering, light, diseases, or how to earn points!'])
]

# Create the chatbot
chatbot = Chat(patterns, reflections)

# Function to start the conversation (for testing in Colab/Terminal)
def start_chat():
    print("OnlyPlants Bot: Hello! I'm your smart orchid assistant. How can I help you today? (type 'quit' to exit)")
    while True:
        user_input = input("You: ")
        if user_input.lower() == 'quit':
            print("OnlyPlants Bot: Goodbye! Remember to keep your orchids happy!")
            break
        response = chatbot.respond(user_input)
        print("OnlyPlants Bot:", response)

# Uncomment the line below to run the chat loop when executing this script directly
start_chat()
```
